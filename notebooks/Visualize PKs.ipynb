```python
import pandas as pd
import seaborn as sns

knotty = pd.read_csv('/home/gnye8/Desktop/PK_research/pipeline_results/knotty/knotty_analysis_scores.csv')

pknots = pd.read_csv('/home/gnye8/Desktop/PK_research/pipeline_results/pknots/pknots_analysis_scores.csv')
```

```python
knotty.sort_values('shape_score', ascending=False)
```

```python
pknots.sort_values('shape_score', ascending=False)
```

```python
# plotting: shape vs. pk bp shape score
# shape score vs. rank
# shape score vs. consensus

g = sns.jointplot(x="shape_score", y="pk_bp_shape_score", data=knotty, kind="reg", xlim=(0,1.0), ylim=(0,1.0), truncate=False)
g = sns.jointplot(x='shape_score', y='pk_bp_shape_score', data=pknots, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
```

```python
# shape score vs. rank

g = sns.jointplot(x='pk_bp_shape_score', y='rank', data=knotty, kind='reg', xlim=(0,1.0), ylim=(-3,37), truncate=False)
g = sns.jointplot(x='pk_bp_shape_score', y='rank', data=pknots, kind='reg', xlim=(0,1.0), ylim=(-3,37), truncate=False)
```

```python
# shape score vs. consensus

g = sns.jointplot(x='pk_bp_shape_score', y='pk_bp_consensus_score', data=knotty, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
g = sns.jointplot(x='pk_bp_shape_score', y='pk_bp_consensus_score', data=pknots, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
```

```python
g = sns.jointplot(x='shape_score', y='consensus_score', data=knotty, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
g = sns.jointplot(x='shape_score', y='consensus_score', data=pknots, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
```

```python
# rank vs. consensus

g = sns.jointplot(x='consensus_score', y='pk_bp_consensus_score', data=knotty, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
g = sns.jointplot(x='consensus_score', y='pk_bp_consensus_score', data=pknots, kind='reg', xlim=(0,1.0), ylim=(0,1.0), truncate=False)
```

```python
# thresholds: 
# overall shape score: 0.7-0.75
# pk_bp_shape score: 0.8
# rank: 25 (knotty), 20 (pknots)
# consensus: 0.6
# pk_bp consensus: 0.6

```

```python
df = knotty[knotty['shape_score'] > 0.6]
df1 = df[df['pk_bp_shape_score'] > 0.6]
df2 = df1[df1['rank'] > 10]
df3 = df2[df2['consensus_score'] > 0.5]
df4 = df3[df3['pk_bp_consensus_score'] > 0.5]

df4.to_csv('/home/gnye8/Desktop/PK_research/pipeline_results/knotty/knotty_top_results.csv')
df4
```

```python
df5 = pknots[pknots['shape_score']>0.6]
df6 = df5[df5['pk_bp_shape_score']>0.6]
df7 = df6[df6['rank']>10]
df8 = df7[df7['consensus_score']>0.5]
df9 = df8[df8['pk_bp_consensus_score']>0.5]

df9.to_csv('/home/gnye8/Desktop/PK_research/pipeline_results/pknots/pknots_top_results.csv')
df9
```
