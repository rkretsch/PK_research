```python
def get_list(csv):
    csv_locs = csv[['start', 'end']]
    csv_locs_list = csv_locs.values.tolist()
    return csv_locs_list

def get_consensus(base_csv, compare_csv):
    base_locs = get_list(base_csv)
   # print(base_locs)
    compare_locs = get_list(compare_csv)
   # print(compare_locs)
    consensus = []
    for comp_wind in compare_locs:
        for base_wind in base_locs:
            if (comp_wind[0] >= base_wind[0]) and (comp_wind[1] <= base_wind[1]):
                consensus.append(comp_wind)
    final_consensus = []
    for base_wind in base_locs:
        cons_start = []
        cons_end = []
        for cons_wind in consensus:
            if cons_wind[0] >= base_wind[0]:
                cons_start.append(cons_wind[0])
            if cons_wind[1] <= base_wind[1]:
                cons_end.append(cons_wind[1])
        final_consensus.append(min(cons_start))
        final_consensus.append(max(cons_end))
    return final_consensus
```

```python
knotty_consensus = get_consensus(threshknot, knotty)
pknots_consensus = get_consensus(threshknot, pknots)
```
